```python
import cv2
import os
import h5py
import numpy as np
from PIL import Image
from skimage import io
```

```python
TRAIN_IMG = 'data/training/images/'
TRAIN_LABEL = 'data/training/labels/'

TEST_IMG = 'data/test/images/'
TEST_LABEL = 'data/test/labels/'
```

```python
images_train = 'data/training/training/images'
labels_train = 'data/training/training/1st_manual'

Nimgs = 20
channels = 1
height = 584
width = 565

imgs = np.empty((Nimgs,height,width,channels))
lbls = np.empty((Nimgs,height,width))

i=0
for file in sorted(os.listdir(images_train)):
  img = cv2.imread(images_train+'/'+file, 0)
  img = np.expand_dims(img, axis=0)
  img = np.moveaxis(img,0, -1)
  imgs[i] = img
  i += 1

i=0
for file in sorted(os.listdir(labels_train)):
  g_truth = Image.open(labels_train +'/'+ file)
  lbls[i] = np.asarray(g_truth)
  i += 1
```

```python
print(imgs.shape)
print(lbls.shape)
```

```python
for i in range(20):
  res = cv2.imwrite(TRAIN_IMG + str(i) + '.png', imgs[i])
  print(res)

for i in range(20):
  res = cv2.imwrite(TRAIN_LABEL + str(i) + '.png', lbls[i])
  print(res)
```

```python
images_test = 'data/test/test/images'
labels_test = 'data/test/test/1st_manual'

Nimgs = 20
channels = 1
height = 584
width = 565

imgs = np.empty((Nimgs,height,width,channels))
lbls = np.empty((Nimgs,height,width))

i=0
for file in sorted(os.listdir(images_test)):
  img = cv2.imread(images_test+'/'+file, 0)
  img = np.expand_dims(img, axis=0)
  img = np.moveaxis(img,0, -1)
  #print(img.shape)
  imgs[i] = img
  i += 1

i=0
for file in sorted(os.listdir(labels_test)):
  g_truth = Image.open(labels_test +'/'+ file)
  lbls[i] = np.asarray(g_truth)
  i += 1
```

```python
print(imgs.shape)
print(lbls.shape)
```

```python
j = 1
for i in range(20):
  cv2.imwrite(TEST_IMG + str(j) + '.png',imgs[i])
  j += 1

j = 1
for i in range(20):
  cv2.imwrite(TEST_LABEL + str(j) + '.png',lbls[i])
  j += 1
```
